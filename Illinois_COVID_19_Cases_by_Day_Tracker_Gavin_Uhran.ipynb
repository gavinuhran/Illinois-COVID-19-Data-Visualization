```
import requests, json, matplotlib.pyplot as plt, numpy as np

data = json.loads(requests.get('https://www.dph.illinois.gov/sitefiles/COVIDHistoricalTestResults.json?nocache=1').text)
stateTestingResults = data["state_testing_results"]["values"]

changeInCasesByDay = [0]
dates = [stateTestingResults[0]["testDate"]]

for i in range(1, len(stateTestingResults)):
  dates.append(stateTestingResults[i]["testDate"])
  changeInCases = stateTestingResults[i]["confirmed_cases"] - stateTestingResults[i-1]["confirmed_cases"]
  changeInCasesByDay.append(changeInCases)

plt.figure(figsize=(15, 12))
line1, = plt.plot(dates, changeInCasesByDay, label="Actual")
plt.xlabel('Date')
plt.ylabel('Number of COVID-19 Cases')
plt.title('Change in COVID-19 Cases in Illinois by Day')
plt.xticks(rotation=45)
plt.show()
```
